```python
import matminer
from matminer.featurizers.base import MultipleFeaturizer
from matminer.featurizers import composition as cf
from matminer.featurizers.conversions import StrToComposition
import pandas as pd
```

```python
MultipleFeaturizer([cf.ElementProperty(data_source = "magpie",
                                           features = [
                                                        "Number",
                                                        "MendeleevNumber",
                                                        "MeltingT",
                                                        "Column",
                                                        "Row",
                                                        "CovalentRadius",
                                                        "Electronegativity",
                                                        "NsValence",
                                                        "NpValence",
                                                        "NdValence",
                                                        "NfValence",
                                                        "NValence",
                                                        "NsUnfilled",
                                                        "NpUnfilled",
                                                        "NdUnfilled",
                                                        "NfUnfilled",
                                                        "NUnfilled",
                                                        "GSvolume_pa",
                                                       "bulk_modulus"
                                                    ],
                                       stats=["minimum", "maximum","mean", "std_dev"]),
                   cf.ElementProperty(data_source = "pymatgen",
                                           features = [
                                                        "block",
                                                        "atomic_mass",
                                                        "atomic_radius",
                                                        "thermal_conductivity",
                                                        "bulk_modulus",
                                                    ],
                                       stats=["minimum", "maximum","mean", "std_dev"])
                   ]
                    ).feature_labels()
```

```python
"""
需要输入材料的化学式(例如Nb1Pt1Si1)，可以得到材料的145维向量表示

MultipleFeaturizer: 特征组合器
cf.Stoichiometry(): 计算化学计量属性的规范
cf.ElementProperty.from_preset("magpie"): 基于元素信息生成特征
cf.ValenceOrbital(props=['avg']): 计算轨道价电子数avg
cf.IonProperty(fast=True): 判断合成possible
"""

def magpie(df,composition):
    df = StrToComposition(target_col_id='composition_obj').featurize_dataframe(df, composition,ignore_errors=True)
    feature_calculators = MultipleFeaturizer([cf.ElementProperty.from_preset("magpie"),
                                              cf.ValenceOrbital(props=['avg']), cf.IonProperty(fast=True)])

    feature_labels = feature_calculators.feature_labels()

    df = feature_calculators.featurize_dataframe(df, col_id='composition_obj',ignore_errors=True)

    return df
```

```python

def matminer(df,composition):
    df = StrToComposition(target_col_id='composition_obj').featurize_dataframe(df, composition,ignore_errors=True)
    feature_calculators = MultipleFeaturizer([cf.ElementProperty.from_preset("magpie"),
                                              cf.ElementProperty.from_preset("matminer"),])

    feature_labels = feature_calculators.feature_labels()

    df = feature_calculators.featurize_dataframe(df, col_id='composition_obj',ignore_errors=True)

    return df
```

```python
def multipleFeaturizer(df,composition):
    df = StrToComposition(target_col_id='composition_obj').featurize_dataframe(df, composition,ignore_errors=True)
    feature_calculators = MultipleFeaturizer([cf.ElementProperty(data_source = "magpie",
                                           features = [
                                                        "Number",
                                                        "MendeleevNumber",
                                                        "MeltingT",
                                                        "Column",
                                                        "Row",
                                                        "CovalentRadius",
                                                        "Electronegativity",
                                                        "NsValence",
                                                        "NpValence",
                                                        "NdValence",
                                                        "NfValence",
                                                        "NValence",
                                                        "NsUnfilled",
                                                        "NpUnfilled",
                                                        "NdUnfilled",
                                                        "NfUnfilled",
                                                        "NUnfilled",
                                                        "GSvolume_pa",
                                                    ], stats=["minimum", "maximum","mean", "std_dev"]),
                   cf.ElementProperty(data_source = "pymatgen",
                                           features = [
                                                        "block",
                                                        "atomic_mass",
                                                        "atomic_radius",
                                                        "thermal_conductivity",
                                                    ],
                                       stats=["minimum", "maximum","mean", "std_dev"])
                   ]
                    )

    feature_labels = feature_calculators.feature_labels()

    df = feature_calculators.featurize_dataframe(df, col_id='composition_obj',ignore_errors=True)

    return df
```
